```python
# Import Libraries
from pathlib import Path
import numpy_financial as npf
import csv
# Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('menu_data.csv')
sales_filepath = Path('sales_data.csv')
# Initialize list objects to hold our menu and sales data
menu = []
sales = []
# Reads CSV files
with open(menu_filepath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    for row in csvreader:
        row[3] = float(row[3])
        row[4] = int(row[4])
        menu.append(row)
with open(sales_filepath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    for row in csvreader:
        row[3] = int(row[3])
        sales.append(row)
# Create lists of all the columns in menu
menu_items = []
menu_categories = []
menu_description = []
menu_price = []
menu_cost = []
for row in menu:
    menu_items.append(row[0])
    menu_categories.append(row[1])
    menu_description.append(row[2])
    menu_price.append(float(row[3]))
    menu_cost.append(float(row[4]))
# Create lists for all the columns in sales
sales_itemID = []
sales_date = []
sales_cc = []
sales_quantity = []
sales_item = []
for row in sales:
    sales_itemID.append(row[0])
    sales_date.append(row[1])
    sales_cc.append(row[2])
    sales_quantity.append(int(row[3]))
    sales_item.append(row[4])
#Initialize report
report = {}
#Loop through sales and create dictionaries for each menu item
for i, sales_row in enumerate(sales_item):
    for j, menu_row in enumerate(menu_items):
        if sales_row == menu_row:
            item = (sales_item[i])
            if item not in report:
                report[item] = {'01-count':0,
                               '02-revenue':0,
                               '03-cogs':0,
                               '04-profit':0}
            report[item]['01-count'] += sales_quantity[i]
            report[item]['02-revenue'] += menu_price[j] * sales_quantity[i]
            report[item]['03-cogs'] += menu_cost[j] * sales_quantity[i]
            report[item]['04-profit'] += (menu_price[j]-menu_cost[j]) * sales_quantity[i]
# Write output to txt file using csvwriter
output_path = Path('output.txt')
with open(output_path, 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    for name in report:
        csvwriter.writerow(
            [
                name, report[name]
            ]
        )
```
